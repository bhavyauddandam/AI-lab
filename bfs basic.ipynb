```python
# Graph adjacency list representation
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': [],
    'D': ['G', 'H'],
    'E': ['I', 'J'],
    'F': ['K'],
    'G': [],
    'H': ['L'],
    'I': ['N'],
    'J': [],
    'K': [],
    'L': ['M', 'O'],
    'M': [],
    'N': ['P', 'Q', 'R'],
    'P': ['S', 'T'],
    'Q': ['I'],
    'R': ['U', 'V'],
    'S': ['W', 'X'],
    'T': [],
    'U': ['Y'],
    'V': ['Z'],
    'W': [],
    'X': [],
    'Y': [],
    'Z': []
}

# Input start and goal states
start = input("Enter initial state: ").upper()
goal = input("Enter final state: ").upper()

print("\nInitial state:", start)
print("Initial level: 0")

level = 0
paths = [[start]]

# BFS Algorithm with a depth limit of 8
while level <= 8:
    new_paths = []
    for path in paths:
        node = path[-1]

        if node == goal:
            print("Final state:", goal)
            print("Final level:", level)
            print("Path:")
            print(" -> ".join(path))
            exit() # Stop the program once goal is found

        # Expand children
        for child in graph.get(node, []):
            new_paths.append(path + [child])

    paths = new_paths
    level = level + 1

print("Goal not found within 8 levels")
```
