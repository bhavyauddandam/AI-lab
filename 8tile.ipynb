```python
import heapq

goal = [[1,2,3],
        [4,5,6],
        [7,8,0]]

moves = [(-1,0,"Up"),
         (1,0,"Down"),
         (0,-1,"Left"),
         (0,1,"Right")]

# Manhattan Distance Heuristic
def h(state):
    distance = 0
    for i in range(3):
        for j in range(3):
            value = state[i][j]
            if value != 0:
                x = (value-1)//3
                y = (value-1)%3
                distance += abs(i-x) + abs(j-y)
    return distance

# Find blank tile
def blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i,j
    return None

def astar(start):
    pq = [(h(start),0,start,[])]
    visited=set()

    while pq:
        f,g,state,path = heapq.heappop(pq)

        if state == goal:
            for step,(move,st) in enumerate(path,1):
                print("Step",step,"Move:",move)
                for r in st:
                    print(r)
                print()
            print("Goal State Reached!")
            return

        visited.add(str(state))

        pos = blank(state)
        if pos is None:
            print("Error: No blank tile (0) found!")
            return

        x,y = pos

        for dx,dy,move in moves:
            nx,ny = x+dx,y+dy

            if 0<=nx<3 and 0<=ny<3:
                new_state=[row[:] for row in state]
                new_state[x][y],new_state[nx][ny]=new_state[nx][ny],new_state[x][y]

                if str(new_state) not in visited:
                    heapq.heappush(
                        pq,
                        (g+1+h(new_state),g+1,new_state,path+[(move,new_state)])
                    )

print("Enter initial puzzle (use 0 for blank):")

start=[list(map(int,input().split())) for _ in range(3)]

astar(start)
```
