```python
import heapq

goal = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
moves = [(-1, 0, "Up"), (1, 0, "Down"), (0, -1, "Left"), (0, 1, "Right")]


def h(s):
    d = 0
    for i in range(3):
        for j in range(3):
            v = s[i][j]
            if v:
                x, y = (v - 1) // 3, (v - 1) % 3
                d += abs(i - x) + abs(j - y)
    return d


def blank(s):
    for i in range(3):
        for j in range(3):
            if s[i][j] == 0: return i, j


def astar(start):
    pq = [(h(start), 0, start, [])]
    v = set()
    while pq:
        f, g, s, p = heapq.heappop(pq)
        if s == goal:
            for i, (m, st) in enumerate(p, 1):
                print("Step", i, "Move:", m)
                for r in st:
                    print(r)
                print()
            print("Goal State Reached!")
            return
        v.add(str(s))
        x, y = blank(s)
        for dx, dy, m in moves:
            nx, ny = x + dx, y + dy
            if 0 <= nx < 3 and 0 <= ny < 3:
                ns = [r[:] for r in s]
                ns[x][y], ns[nx][ny] = ns[nx][ny], ns[x][y]
                if str(ns) not in v:
                    heapq.heappush(pq,
                                   (g + 1 + h(ns), g + 1, ns, p + [(m, ns)]))


print("Enter initial puzzle (use 0 for blank):")

start = [list(map(int, input().split())) for _ in range(3)]
astar(start)
```
