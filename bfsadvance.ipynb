```python
from collections import deque

graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': [],
    'D': ['I', 'J'],
    'E': ['G', 'H'],
    'F': [],
    'G': [],
    'H': [],
    'I': ['K', 'L'],
    'J': [],
    'K': [],
    'L': []
}

SOURCE = 'A'
GOAL = 'L'

# ---------- PRINT TREE ----------

def print_tree_bfs(graph, source):
    queue = deque([(source, 0)])
    visited = set()

    while queue:
        node, level = queue.popleft()
        if node not in visited:
            visited.add(node)
            print("  " * level + node)
            for child in graph.get(node, []):
                queue.append((child, level + 1))


# ---------- BFS TRAVERSAL ----------

def bfs_traversal(graph, source):
    visited = []
    queue = deque([source])

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.append(node)
            queue.extend(graph[node])

    return visited


# ---------- BFS PATH ----------

def bfs_path(graph, start, goal):
    queue = deque([[start]])
    visited = set()

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node not in visited:
            visited.add(node)

            if node == goal:
                return path

            for n in graph[node]:
                queue.append(path + [n])

    return None


# ---------- MAIN OUTPUT ----------

def bfs_full_output(graph):
    all_nodes = set(graph.keys())
    bfs_order = bfs_traversal(graph, SOURCE)

    print("\nINPUT STATE SPACE TREE (BFS):\n")
    print_tree_bfs(graph, SOURCE)

    print("\n" + "=" * 70)
    print("BFS DETAILED OUTPUT\n")

    for node in graph:
        print("-" * 60)
        print(f"NODE {node}")

        print("Vertices connected :", graph[node])
        print("Edges formed       :", [(node, v) for v in graph[node]])

        idx = bfs_order.index(node) + 1
        visited_until = bfs_order[:idx]
        non_visited = list(all_nodes - set(visited_until))

        print("Visited nodes      :", visited_until)
        print("Non-visited nodes  :", non_visited)

        path = bfs_path(graph, node, GOAL)
        if path:
            print("Path to goal L     :", " -> ".join(path))
        else:
            print("Path to goal L     : No path")

    optimal_path = bfs_path(graph, SOURCE, GOAL)

    print("\n" + "=" * 70)
    print("OPTIMAL SOLUTION PATH:")
    print(" -> ".join(optimal_path))

    # Modified graph (optimal path only)
    modified_graph = {}
    for i in range(len(optimal_path) - 1):
        modified_graph[optimal_path[i]] = [optimal_path[i + 1]]
    modified_graph[optimal_path[-1]] = []

    print("\nMODIFIED STATE SPACE TREE (BFS):\n")
    print_tree_bfs(modified_graph, SOURCE)


# ---------- RUN ----------

bfs_full_output(graph)
```
